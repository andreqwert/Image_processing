# Часть 1. Чтение и запись изображений

```python
%matplotlib inline
```

```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-color.png')
```

```python
imshow(img)
```

```python
# Форма изображения
img.shape # Выдают тройку чисел: число строк, число столбцов, количество цветовых каналов
```

```python
# если требуется количество столбцов изображения
img.shape[1]
```

```python
# Посмотрим на какой-то конкретный пиксель
img[383, 374, 2] # в ответ получили число, кодирующее цвет (от 0 до 255)
```

```python
img[383, 374] = [255, 255, 0] # изменили цвет пикселя (383, 374) на жёлтый
```

```python
imshow(img)
```

```python
imsave('tiger-yellow-nose.png', img)
```

# Часть 2. Чтение, запись, работа с пикселями

```python
img = imread('tiger-color.png')
# Выделение прямоугольной области
nose = img[370:410, 350:440] #записана прямоугольная часть изображения. 
# с 370 по 410 строчки, 410 строка не включается; с 350 по 440 стобцы, 440 столбец не включается.
```

```python
imshow(nose)
```

```python
# Поменяем область с носом тигра на малиновый цвет
img[370:410, 350:440] = [255, 0, 255]
imshow(img)
```

```python
# Посмотрим, как мы поменяли пиксель с оранжевого на желтый
img_yellow = imread('tiger-yellow-nose.png')
imshow(img_yellow[370:410, 350:440])
```

```python
img_assigned = img_yellow
```

```python
img_copy = img_yellow.copy() 
```

```python
img_yellow[370:410, 350:440] = [255, 0, 255]
```

```python
imshow(img_yellow)
```

```python
imshow(img_assigned) # несмотря на то, что переменных две (img_yellow, img_assigned), они ссылаются на одно и то же изображение
```
